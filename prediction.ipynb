```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
model = load_model("model.h5")

with open("scaler.pkl",'rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore' :600,
    'Geography' :'France',	
    'Gender' :'Male',	
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,	
    'EstimatedSalary': 50000
}
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df["Geography"] = pd.Categorical(
    input_df["Geography"], 
    categories=["France", "Germany", "Spain"]
)
input_df
```

```python
le = LabelEncoder()

le.fit(["Female", "Male"])

input_df["Gender"] = le.transform(input_df["Gender"])
```

```python
input_df = pd.get_dummies(input_df,columns=["Geography"], dtype=int)
```

```python
input_df
```

```python
scaled_input = scaler.transform(input_df)
```

```python
scaled_input
```

```python
predict_prob = model.predict(scaled_input)
predict_prob
```

```python
if predict_prob>0.5:
    print('The customer is likely to churn')
else:
    print('The customer is not likely to churn')
```
